```python
#2018/ICT/03
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics


data = pd.read_csv("Advertising.csv")
# Part (a)
X_simple = data['TV'].values.reshape(-1,1)  
Y_simple = data['Sales'].values.reshape(-1,1) 
```

```python
X_train_simple, X_test_simple, Y_train_simple, Y_test_simple = train_test_split(X_simple, Y_simple, test_size=0.2, random_state=42)

simple_model = LinearRegression()
simple_model.fit(X_train_simple, Y_train_simple)


print("Intercept (Simple Linear Regression):", simple_model.intercept_)
print("Coefficient (Simple Linear Regression):", simple_model.coef_[0])


predicted_simple = simple_model.predict(X_test_simple)
actual_and_predicted_simple = pd.DataFrame({'Actual': Y_test_simple.flatten(), 'Predicted': predicted_simple.flatten()})
print("\nActual and Predicted Values (Simple Linear Regression):\n", actual_and_predicted_simple.head())

plt.scatter(X_test_simple, Y_test_simple, color='blue')
plt.plot(X_test_simple, predicted_simple, color='red', linewidth=2)
plt.title('Simple Linear Regression')
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')
plt.show()


r_squared_simple = metrics.r2_score(Y_test_simple, predicted_simple)
mae_simple = metrics.mean_absolute_error(Y_test_simple, predicted_simple)
mse_simple = metrics.mean_squared_error(Y_test_simple, predicted_simple)
rms_simple = np.sqrt(mse_simple)

print("\nPerformance Metrics (Simple Linear Regression):")
print("R squared:", r_squared_simple)
print("Mean Absolute Error:", mae_simple)
print("Mean Squared Error:", mse_simple)
print("Root Mean Squared Error:", rms_simple)


# Part (b)
X_multiple = data[['TV', 'Radio', 'Newspaper']] 
Y_multiple = data['Sales']  


X_train_multiple, X_test_multiple, Y_train_multiple, Y_test_multiple = train_test_split(X_multiple, Y_multiple, test_size=0.2, random_state=42)


multiple_model = LinearRegression()
multiple_model.fit(X_train_multiple, Y_train_multiple)


print("\nIntercept (Multiple Linear Regression):", multiple_model.intercept_)
print("Coefficients (Multiple Linear Regression):", multiple_model.coef_)


predicted_multiple = multiple_model.predict(X_test_multiple)
actual_and_predicted_multiple = pd.DataFrame({'Actual': Y_test_multiple.values, 'Predicted': predicted_multiple})
print("\nActual and Predicted Values (Multiple Linear Regression):\n", actual_and_predicted_multiple.head())


r_squared_multiple = metrics.r2_score(Y_test_multiple, predicted_multiple)
mae_multiple = metrics.mean_absolute_error(Y_test_multiple, predicted_multiple)
mse_multiple = metrics.mean_squared_error(Y_test_multiple, predicted_multiple)
rms_multiple = np.sqrt(mse_multiple)

print("\nPerformance Metrics (Multiple Linear Regression):")
print("R squared:", r_squared_multiple)
print("Mean Absolute Error:", mae_multiple)
print("Mean Squared Error:", mse_multiple)
print("Root Mean Squared Error:", rms_multiple)
```
